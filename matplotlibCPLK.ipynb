```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2,3,4,5,6],[2,6,8,10,16])
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("First Graph")
```

# **Subplot**

```python
x = np.arange(1, 10, 0.1)
y1 = x
y2 = np.sqrt(x)
y3 = np.power(x, 2)
y4 = np.power(x, 3)
plt.figure()

plt.subplot(2,2,1)
plt.plot(x, y1, color="c", linestyle='-', marker=".")
plt.title("y1 = x")

plt.subplot(2,2,2)
plt.plot(x, y2, color="k", linestyle='-', marker=".")
plt.title("$y2 = \sqrt{x}$")

plt.subplot(2,2,3)
plt.plot(x, y3, color="b", linestyle='-', marker=".")
plt.title("$y3 = x^2$")

plt.subplot(2,2,4)
plt.plot(x, y4, color="r", linestyle='-', marker=".")
plt.title("$y4 = x^3$")

plt.show()
```

# **Histogram**

```python
marks = [35, 65, 34, 20, 78, 80, 98, 68, 48, 59, 45, 65]
plt.hist(marks)
plt.show()
```

```python
marks = [35, 65, 34, 20, 78, 80, 98, 68, 48, 59, 45, 65]
plt.hist(marks, bins=4)
plt.show()
```

```python
marks = [35, 65, 34, 20, 78, 80, 98, 68, 48, 59, 45, 65]
plt.hist(marks, bins=[0, 25, 50, 75, 100])
plt.show()
```

```python
marks = [35, 65, 34, 20, 78, 80, 98, 68, 48, 59, 45, 65]
plt.hist(marks, bins=[0, 25, 50, 75, 100], rwidth = 0.7)
plt.show()
```

```python
marks = [35, 65, 34, 20, 78, 80, 98, 68, 48, 59, 45, 65]
plt.hist(marks, bins=[0, 25, 50, 75, 100], rwidth = 0.7, color="green")
plt.show()
```

## **Pie chart**

```python
category = ["food", "transport", "rent", "other"]
cost = [1200, 500, 800, 600]
plt.pie(cost, labels= category, radius = 0.7, autopct="%0.1f%%", explode=[0,0.3,0,0])
plt.show()
```
